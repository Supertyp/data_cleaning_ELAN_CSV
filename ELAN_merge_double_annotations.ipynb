```python
import os
from pympi import Eaf
import xml.etree.ElementTree as ET


input_folder = 'C:/Users/barth/Documents/LDACA\data_cleaning/AAE/Transcripts_Anonymised'
#input_folder = 'C:/Users/barth/Documents/LDACA\data_cleaning/AAE/test'

def handle_file(file_path):
    
    annotation_old = "xxx"
    annotation_text_old = "xxx"
    print (file)
    
    
    # Load the ELAN file
    tree = ET.parse(file_path)
    root = tree.getroot()

    for tier in root.findall(".//TIER"):
        
        # iterate over every annotation, comparing the text with the text of the previous annotation
        # if old text = new text: change old annotation final time stamp and delete new annotation
        for annotation in tier.findall(".//ANNOTATION"):
            
            alignable = annotation.find(".//ALIGNABLE_ANNOTATION")
            newTS2 = alignable.attrib['TIME_SLOT_REF2']
    
            
            annotation_values = annotation.findall(".//ANNOTATION_VALUE")
            
            text = annotation_values[0].text

            if text and (text == annotation_text_old) and (len(text.split()) > 10):

                toChangeAlignable = annotation_old.find(".//ALIGNABLE_ANNOTATION")

                toChangeAlignable.attrib['TIME_SLOT_REF2'] = newTS2


                for element in root.iter():
                    for child in element:
                        for grandchild in child:
                            if grandchild is alignable:
                                element.remove(child)

            else:


                annotation_text_old = text

                annotation_old = annotation

        
    with open(file_path, 'wb') as f:
        tree.write(f, encoding='utf-8', xml_declaration=True)
    
                
for root, dirs, files in os.walk(input_folder):
        for file in files:
            if file.endswith('.eaf'):
                file_path = os.path.join(input_folder, file)
                handle_file(file_path)
                
```
